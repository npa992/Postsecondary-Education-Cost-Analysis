```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(pd.read_csv('delta_public_00_12.csv',encoding='cp1252'))
df.head()
```

```python
revenue_df=df[['academicyear','unitid','instname','sector_revised','control',
               'hospital','cpi_scalar_2012','hepi_scalar_2012','fte_count',
              'tuition03','net_student_tuition','federal03','state_local_app',
               'federal07','state_local_grant_contract','federal10','state09','fed_state_loc_grants_con',
               'private03','affiliate01','investment01','priv_invest_endow','auxiliary03','hospital03',
               'other03','other05','stable_operating_rev','total03_revenue','tuition_reliance_a1',
               'tuition_reliance_c2','govt_reliance_a','govt_reliance_c','appliedaid01','grant01','grant02',
               'grant03','grant04','grant05','grant06',
               'grant07','institutional_grant_aid','any_aid_pct','fed_grant_pct',
                'fed_grant_avg_amount','state_grant_pct','state_grant_avg_amount','inst_grant_pct',
               'inst_grant_avg_amount','loan_pct','loan_avg_amount','totalcompletions']]
```

```python
expenses_df=df[['academicyear','instname','sector_revised','cpi_scalar_2012','hepi_scalar_2012','fte_count',
                 'matched_n_02_12_11','landgrnt','unitid',
                 'instruction01','instruction02','research01','research02','pubserv01','pubserv02',
                 'acadsupp01','acadsupp02','studserv01','studserv02','instsupp01','instsupp02','opermain01',
                 'opermain02','depreciation01','grants01','auxiliary01','auxiliary02','hospital01','hospital02',
                'independ01','independ02','otheroper01','otheroper02','totaloper01','totaloper02','other01',
                'other02','totalnon01','totalnon02','total01','total02','total06','eandg01','eandg01_sum',
                'eandg01_w_auxother_sum','eandg02','eandg04','rschpub01','acadinststud01','acadinstsupp01',
                'education_share','noneducation_share','other_ed_related_cost','instruction_share','studserv_share',
                'admin_share','eandr','eandr_degree','eandr_completion','research_share','research_related_cost',
                'pubserv_share','pubserv_related_cost','auxother_cost','sticker_subsidy','average_subsidy',
                'sticker_price_share','nettuition_share','average_subsidy_share','gross_operating_margin',
                'instr_sal_as_pct_instrtot','labor_share_of_instructcost','research_sal_as_pct_restot',
                'labor_share_of_rescost','acadsupp_sal_as_pct_acadsupptot','labor_share_of_acadsuppcost',
                'studserv_sal_as_pct_studservtot','labor_share_of_studservcost','instsupp_sal_as_pct_instsupptot',
                'labor_share_of_instsuppcost','pubserv_sal_as_pct_pubservtot','labor_share_of_pubservcost',
                'assets06','liabilities07','assets11','land04','buildings05','equipment05','endow02m',
                 'totalcompletions']]
df1=expenses_df.loc[(expenses_df['sector_revised']==1)|(expenses_df['sector_revised']==2)]
df1.count()
```

```python
gb=df1.groupby('unitid')
df2=gb.count()
df3=df2.loc[
    (df2['fte_count']==13)&
    (df2['instruction01']==13)&
    (df2['research01']==13)&
    (df2['pubserv01']==13)&
    (df2['acadsupp01']==13)&
    (df2['studserv01']==13)&
    (df2['instsupp01']==13)&
    (df2['auxiliary01']==13)
]
df4=df1.set_index('unitid')
df5=df4.loc[df3.index]

df5['instruction per student']=df5['instruction01']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['research per student']=df5['research01']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['pubserv per student']=df5['pubserv01']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['acadsupp per student']=df5['acadinstsupp01']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['studserv per student']=df5['studserv01']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['instsupp per student']=df5['instsupp01']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['auxiliary per student']=df5['auxiliary01']/(df5['fte_count']*df5['cpi_scalar_2012'])



gb2=df5.groupby('academicyear')
df6=gb2.mean().reset_index()
```

```python
df6.plot(x='academicyear',y='auxiliary per student')
```

```python
df1=revenue_df.loc[(revenue_df['control']==2)]
df1.count()
```

```python
revenue_nonprofit=df1[['academicyear','unitid','instname','sector_revised',
               'hospital','cpi_scalar_2012','hepi_scalar_2012','fte_count','stable_operating_rev',        
              'tuition03','net_student_tuition','federal07','federal10','state09',
               'private03','affiliate01','investment01','auxiliary03','hospital03',
               'other03','other05','appliedaid01','grant01','grant02',
               'grant03','grant07','institutional_grant_aid','any_aid_pct','fed_grant_pct',
                'fed_grant_avg_amount','state_grant_pct','state_grant_avg_amount','inst_grant_pct',
               'inst_grant_avg_amount','loan_pct','loan_avg_amount','totalcompletions']]

gb=revenue_nonprofit.groupby('unitid')
df2=gb.count()
df3=df2.loc[
    (df2['fte_count']==13)&
    (df2['private03']==13)&
    (df2['tuition03']==13)&
    (df2['net_student_tuition']==13)&
    (df2['federal10']==13)&
    (df2['state09']==13)&
    (df2['private03']==13)&
    (df2['investment01'])&
    (df2['other03'])&
    (df2['other05'])&
    (df2['grant01']==13)&
    (df2['grant03']==13)&
    (df2['grant07']==13)&
    (df2['institutional_grant_aid']==13)

]
df4=revenue_nonprofit.set_index('unitid')
df5=df4[['instname','fte_count']].loc[df3.index]
gb2=df5.groupby('instname')
gb2.count()
```

```python
revenue_4y_uni=df1[['academicyear','unitid','instname','sector_revised',
               'hospital','cpi_scalar_2012','hepi_scalar_2012','fte_count','stable_operating_rev',        
              'tuition03','net_student_tuition','federal07','federal10','state09',
               'private03','affiliate01','investment01','auxiliary03','hospital03',
               'other03','other05','appliedaid01','grant01','grant02',
               'grant03','grant07','institutional_grant_aid','any_aid_pct','fed_grant_pct',
                'fed_grant_avg_amount','state_grant_pct','state_grant_avg_amount','inst_grant_pct',
               'inst_grant_avg_amount','loan_pct','loan_avg_amount','totalcompletions']]
gb=revenue_4y_uni.groupby(['unitid'])
df2=gb.count()
df3=df2.loc[
    (df2['fte_count']==13)&
    (df2['tuition03']==13)&
    (df2['net_student_tuition']==13)&
    (df2['federal10']==13)&
    (df2['state09']==13)&
    (df2['private03']==13)&
    (df2['investment01'])&
    (df2['other03'])&
    (df2['other05'])&
    (df2['grant01']==13)&
    (df2['grant03']==13)&
    (df2['grant07']==13)&
    (df2['institutional_grant_aid']==13)
]
df3.head()
```

```python
df6=df2.loc[
    (df2['fte_count']==13)&(df2['stable_operating_rev']==13)
]
df4=revenue_4y_uni.set_index('unitid')

df7=df4[['instname','cpi_scalar_2012','hepi_scalar_2012',
         'academicyear','fte_count','tuition03','net_student_tuition','federal10','stable_operating_rev',
        'state09','auxiliary03','grant01','grant03','grant07','institutional_grant_aid']].loc[df6.index]

df7['Revenue per student']=df7['stable_operating_rev']/(df7['cpi_scalar_2012']*df7['fte_count'])
gb2=df7.groupby('academicyear')

df8=gb2.mean().reset_index()
```

```python
df4=revenue_4y_uni.set_index('unitid')
df5=df4[['instname','cpi_scalar_2012','hepi_scalar_2012',
         'academicyear','fte_count','tuition03','net_student_tuition','federal10',
        'state09','auxiliary03','grant01','grant03','grant07','institutional_grant_aid']].loc[df3.index]
```

```python
df5['Tuition per student (2012 dollars)']=df5['tuition03']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['Net Tuition per student (2012 dollars)']=df5['net_student_tuition']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['Federal revenue per student (2012 dollars)']=df5['federal10']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['State/Local revenue per student (2012 dollars)']=df5['state09']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['Auxiliary revenue per student (2012 dollars)']=df5['auxiliary03']/(df5['fte_count']*df5['cpi_scalar_2012'])
df5['Institutional grant per student (2012 dollars)']=(df5['institutional_grant_aid']/
                                                    (df5['fte_count']*df5['cpi_scalar_2012']))
df5['Total government expenditure']=df5['Federal revenue per student (2012 dollars)']+df5['State/Local revenue per student (2012 dollars)']
df5['']
```

```python
gb1=df5.groupby(['academicyear'])
complete_revenue_4y_df=gb1.mean().reset_index()
complete_revenue_4y_df.head()
```

```python
complete_revenue_4y_df.plot(x='academicyear',y='Total government expenditure')
```
