```python
# import dependencies

import pandas as pd
```

```python
# load the data into a dataframe

df=pd.DataFrame(pd.read_csv('raw_data/delta_public_00_12.csv',encoding='cp1252', low_memory=False))
df.head()
```

```python
# Locate public and private non-profit universities using the "control" column, which is "1" for public universities
# and "2" for private non-profit universities

df1=df.loc[
    (df['control']==1)|
    (df['control']==2)]
```

```python
# "unitid" in the unique identification number for each university.
# Groupby "unitid" and then do a count to get the number of years each university has reported the data 

gb=df1.groupby('unitid').count().reset_index()
```

```python
# Gather an array of unitid of those universities which have reported the necessary expense data for all 13 years

expense_id=gb['unitid'].loc[
    (gb['fte_count']==13)&
    (gb['instruction01']==13)&
    (gb['research01']==13)&
    (gb['pubserv01']==13)&
    (gb['acadsupp01']==13)&
    (gb['studserv01']==13)&
    (gb['instsupp01']==13)&
    (gb['auxiliary01']==13)&
    (gb['opermain01']==13)&
    (gb['grants01']==13)&
    (gb['eandg01']==13)&
    (gb['eandg01_sum']==13)&
    (gb['eandg01_w_auxother_sum']==13)
]

expense_id.count()
```

```python
# Gather an array of unitid of those universities which have reported the necessary revenue data for all 13 years

revenue_id=gb['unitid'].loc[
    (gb['fte_count']==13)&
    (gb['tuition03']==13)&
    (gb['net_student_tuition']==13)&
    (gb['federal10']==13)&
    (gb['state09']==13)&
    (gb['private03']==13)&
    (gb['auxiliary03']==13)&
    (gb['other05']==13)&
    (gb['grant01']==13)&
    (gb['grant02']==13)&
    (gb['grant03']==13)&
    (gb['grant07']==13)&
    (gb['institutional_grant_aid']==13)]

revenue_id.count()
```

```python
# Gather an array of unitid of those universities which have reported the necessary complete data for all 13 years

complete_id=gb['unitid'].loc[
    (gb['fte_count']==13)&
    (gb['tuition03']==13)&
    (gb['net_student_tuition']==13)&
    (gb['federal10']==13)&
    (gb['state09']==13)&
    (gb['private03']==13)&
    (gb['auxiliary03']==13)&
    (gb['other05']==13)&
    (gb['grant01']==13)&
    (gb['grant02']==13)&
    (gb['grant03']==13)&
    (gb['grant07']==13)&
    (gb['institutional_grant_aid']==13)&
    (gb['instruction01']==13)&
    (gb['research01']==13)&
    (gb['pubserv01']==13)&
    (gb['acadsupp01']==13)&
    (gb['studserv01']==13)&
    (gb['instsupp01']==13)&
    (gb['auxiliary01']==13)&
    (gb['opermain01']==13)&
    (gb['grants01']==13)&
    (gb['eandg01']==13)&
    (gb['eandg01_sum']==13)&
    (gb['eandg01_w_auxother_sum']==13)]

complete_id.count()
```

```python
# Set "unitid" as index on the main "df" so that we can  locate the universities with complete dataset using
# the "unitid" array we got from above

df2=df.set_index('unitid')
```

```python

expense_df=df2[['academicyear','instname','cpi_scalar_2012','hepi_scalar_2012',
                'fte_count','instruction01','research01','pubserv01','acadsupp01',
               'studserv01','instsupp01','auxiliary01','opermain01','grants01','eandg01',
               'eandg01_sum','eandg01_w_auxother_sum']].loc[expense_id]

revenue_df=df2[['academicyear','instname','cpi_scalar_2012','hepi_scalar_2012','fte_count',
                'tuition03','net_student_tuition','federal10','state09',
               'private03','auxiliary03','other05','grant01','grant02',
               'grant03','grant07','institutional_grant_aid']].loc[revenue_id]

complete_df=df2[['academicyear','instname','cpi_scalar_2012','hepi_scalar_2012',
                 'fte_count','tuition03','net_student_tuition','federal10','state09',
                 'private03','auxiliary03','other05','grant01','grant02','grant03',
                 'grant07','institutional_grant_aid','instruction01','research01','pubserv01',
                 'acadsupp01','studserv01','instsupp01','auxiliary01','opermain01','grants01',
                 'eandg01','eandg01_sum','eandg01_w_auxother_sum']].loc[complete_id]

```

```python
# export the resulting df

revenue_df.to_csv('clean_data/revenue_nonprofit.csv')
expense_df.to_csv('clean_data/expenses_nonprofit.csv')
complete_df.to_csv('clean_data/complete_nonprofit.csv')
```
